```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate synthetic data
np.random.seed(42)
X1 = np.random.rand(100, 1) * 10  # Feature 1
X2 = np.random.rand(100, 1) * 5   # Feature 2
y = 3 * X1 + 2 * X2 + np.random.randn(100, 1) * 2  # Target with noise
```

```python
# Combine features into a single array
X = np.hstack((X1, X2))
```

```python
# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predicted values
y_pred = model.predict(X)
```

```python
# Model parameters
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Mean Squared Error (MSE):", mean_squared_error(y, y_pred))
print("R-squared (R^2):", r2_score(y, y_pred))
```

```python
# Visualization in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot actual data
ax.scatter(X1, X2, y, color='blue', label='Actual Data')

# Create a grid for surface plot
X1_grid, X2_grid = np.meshgrid(np.linspace(X1.min(), X1.max(), 10),
                               np.linspace(X2.min(), X2.max(), 10))

# Predict values for the grid
y_grid = model.intercept_ + model.coef_[0][0] * X1_grid + model.coef_[0][1] * X2_grid


# Plot the regression plane
ax.plot_surface(X1_grid, X2_grid, y_grid, color='orange', alpha=0.6, label='Regression Plane')


# Labels and title
ax.set_title("Multiple Linear Regression Visualization")
ax.set_xlabel("Feature 1 (X1)")
ax.set_ylabel("Feature 2 (X2)")
ax.set_zlabel("Target (y)")
ax.legend()

plt.show()
```
